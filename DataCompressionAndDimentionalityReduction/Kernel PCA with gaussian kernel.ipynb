```python
from scipy.spatial.distance import pdist, squareform
from scipy import exp
from scipy.linalg import eigh
import numpy as np
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def rbf_kernel_pca(X, gamma, n_components):
    sq_dist = pdist(X, 'sqeuclidean') #pairwise euclidian distance of M*N dimentional dataset
    
    mat_sq_dist = squareform(sq_dist) #coverts pairwise distance into square matrix
    
    K = exp(-gamma*mat_sq_dist) #Computes the symmetric kernel matrix
    
    N = K.shape[0]
    one_n = np.ones((N,N))/N
    K = K - one_n.dot(K) - K.dot(one_n) + one_n.dot(K).dot(one_n)
    
    eVal, eVec = eigh(K)
    eVal, eVec = eVal[::-1], eVec[::-1]
    
    X_pc = np.column_stack([eVec[:,i] for i in range(n_components)])
    
    return X_pc
```

```python
#Seperating Half Moons
```

```python
from sklearn.datasets import make_moons
X,y  = make_moons(n_samples=100, random_state=123)
plt.scatter(X[y==0,0],X[y==0,1], marker = 's', color = 'red',alpha=0.5)
plt.scatter(X[y==1,0],X[y==1,1], marker = '^', color = 'blue',alpha = 0.5)
plt.tight_layout()
plt.show()
```

```python
#Projecting the dataset to it's prinicpal components using standard pca
```

```python
from sklearn.decomposition import PCA
scikit_PCA = PCA(n_components=2)
X_spca = scikit_PCA.fit_transform(X)
```

```python
fig, ax = plt.subplots(nrows = 1, ncols=2, figsize =(7,3))
ax[0].scatter(X_spca[y==0,0],X_spca[y==0,1], marker = 's', color = 'red',alpha=0.5)
ax[0].scatter(X_spca[y==1,0],X_spca[y==1,1], marker = '^', color = 'blue',alpha = 0.5)
ax[1].scatter(X_spca[y==0,0], np.zeros((50,1))+0.02, marker = 's', color = 'red',alpha=0.5)
ax[1].scatter(X_spca[y==0,1], np.zeros((50,1))-0.02, marker = '^', color = 'blue',alpha = 0.5)
ax[0].set_xlabel('PC1')
ax[0].set_ylabel('PC2')
ax[1].set_yticks([])
ax[1].set_xlabel('PC1')
plt.tight_layout()
plt.show()
```

```python
# As we can see that standard pca is unable to classify the two moon problem
# We will use kernel pca next with radial basos function kernel
```

```python
start, stop, num = 11,12,50
fig, ax = plt.subplots(nrows = len(np.linspace(start, stop, num)), ncols=2, figsize =(10,3*len(np.linspace(start, stop, num))))
j=0
for i in np.linspace(start, stop, num):
    X_kpca = rbf_kernel_pca(X, gamma=i, n_components = 2)
    ax[j,0].scatter(X_kpca[y==0,0],X_kpca[y==0,1], marker = 's', color = 'red',alpha=0.5)
    ax[j,0].scatter(X_kpca[y==1,0],X_kpca[y==1,1], marker = '^', color = 'blue',alpha = 0.5)
    ax[j,1].scatter(X_kpca[y==0,0], np.zeros((50,1))+0.02, marker = 's', color = 'red',alpha=0.5)
    ax[j,1].scatter(X_kpca[y==0,1], np.zeros((50,1))-0.02, marker = '^', color = 'blue',alpha = 0.5)
    #ax[i,0].set_xlabel('PC1')
    ax[j,0].set_ylabel(i)
    ax[j,1].set_yticks([-1,1])
    #ax[i,1].set_xlabel('PC1')
    j+=1
plt.show()
```
